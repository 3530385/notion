```python
import pandas as pd
d={
    'Name':tasks,
    'Due':["October 31, 2021" for _ in tasks],
    'Late':[0 for _ in tasks],
    'Cost':[round(150000/len(tasks),2) for _ in tasks],
    'Time':[None for _ in tasks],
    'Done':['No' for _ in tasks],
    'Status':['Not started' for _ in tasks],
    'Order':[i for i,_ in enumerate(tasks)]
}
df=pd.DataFrame(d)
df.to_csv('gdz9table.csv')
```
